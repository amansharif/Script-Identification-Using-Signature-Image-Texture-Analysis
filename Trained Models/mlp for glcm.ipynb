```python
import pandas as pd
```

```python
df = pd.read_csv('glcm_features.csv')
```

```python
df.head()
```

```python
y = df[['f113']]
X = df.drop('f113', axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 25, shuffle = True)
```

```python
from  sklearn.neural_network import MLPClassifier
```

```python
#s = (100,100,10)
```

```python
mlp_c = MLPClassifier(hidden_layer_sizes=(58), activation='logistic', solver='sgd', alpha=0.001, learning_rate_init=0.0001, max_iter=1000, verbose=False)
```

```python
mlp_c.fit(X_train,y_train)
```

```python
mlp_c.score(X_test,y_test)
```
